```python
# Importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display dataset structure
print("Dataset Shape:", iris.shape)
print("\nColumn Names:", iris.columns.tolist())
print("\nFirst 5 Rows:")
print(iris.head())

# Set the style for plots
sns.set(style='whitegrid')
```

```python
# Scatter Plot: Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set2')
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
# Histogram: Distribution of all numerical columns
iris.hist(figsize=(10, 8), bins=15, edgecolor='black', color='skyblue')
plt.suptitle("Distribution of Iris Features")
plt.tight_layout()
plt.show()
```

```python
# Box Plot: Spread and outliers of all features
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, palette='Set3')
plt.title("Boxplot of Iris Features")
plt.xticks(rotation=45)
plt.show()
```
